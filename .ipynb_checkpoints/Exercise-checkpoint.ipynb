```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('Price_of_icecream.xlsx')
x = df['Year']
y = df['Price']

plt.plot(x, y, 'k.-',markersize=10)
bars = plt.bar(x,y)
for bar in bars:
    height = bar.get_height()
    if height <= 3:
        bar.set_color('green')
    elif height <= 4.5:
        bar.set_color('blue')
    else:
        bar.set_color('red')

plt.title('Price of ice-cream')


plt.show()
```
