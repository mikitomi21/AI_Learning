```python
x = [1,2,3]
y = [2,4,6]
plt.plot(x,y, label='2x')

y1 = [1,2,3]
x1 = [1,2,3]
plt.plot(x1,y1, label='x')

plt.title("Chart title", fontdict={'fontname':'Comic Sans MS', 'fontsize':20})
plt.xlabel("X_Label")
plt.ylabel("Y_Label")

plt.xticks([1,2,3])
plt.yticks([1,2,3,4,5,6])

plt.legend()

plt.show()
```
