```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.arange(0,6.4,0.1)
y = np.sin(x)
plt.plot(x,y)
```

```python
x = [1,2,3]
y = [2,4,6]
plt.plot(x,y, label='2x')

y1 = [1,2,3]
x1 = [1,2,3]
plt.plot(x1,y1, label='x', color="pink", linewidth=3, marker ='.', markersize=10, markeredgecolor="blue", linestyle='--')

plt.title("Chart title", fontdict={'fontname':'Comic Sans MS', 'fontsize':20})
plt.xlabel("X_Label")
plt.ylabel("Y_Label")

plt.xticks([1,2,3])
plt.yticks([1,2,3,4,5,6])

plt.legend()

plt.show()
```

```python
# shorthand notation
# [color][marker][line]

plt.figure(figsize=(5,4), dpi=120)

x = [1,2,3,4,5]
y = [3,2.4,1,3,2]
plt.plot(x,y, 'g.--', label="A1")

x = np.arange(1,5, 0.5)
plt.plot(x, x**2, 'r|:', label="x^2")

plt.legend()

plt.savefig('mygraph.png', dpi=300)
plt.show()
```

```python
x = ['eat', 'sleep', 'programming']
y = [20,10,70]
bars = plt.bar(x,y)

bars[0].set_hatch('/')
bars[1].set_hatch('|')
bars[2].set_hatch('.')
plt.show()
```
